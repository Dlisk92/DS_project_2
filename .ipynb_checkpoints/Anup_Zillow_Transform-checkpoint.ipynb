```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import minmax_scale
%matplotlib inline
```

```python
df = pd.read_csv('Zillow_KC_listings.csv')
```

```python
df.head()
```

```python
df = df.drop(columns = 'Unnamed: 0')
```

```python
log_n_scale = ['sqft_living']
scale = ['bedrooms', 'bathrooms',  'lat', 'long', 'zipcode']

transforms = [log_n_scale, scale]
```

```python
for transform in transforms:
        for col in transform:
            if transform == log_n_scale:
                df[col] = np.log1p(df[col])
                
            df[col] = minmax_scale(df[col])

df.head()
```

```python
df['price'] = np.log1p(df.price)
```

```python
df
```

```python
df.to_csv('Zillow_data_transformed.csv')
```
